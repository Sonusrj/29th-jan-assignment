#### Q2. Create a simple Flask application to display ‘Hello World!!’.World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")

def hello():
    return "Hello world!!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### Q3. What is App routing in Flask? Why do we use app routes?

App routing in flask is technique to help a user to remember application URLs . It is useful to access the desired page directly without having to navigate from the homepage.
the`rout()` decorator in flask is used to bind URL to function .

Why we use:
1) app rout help us organize our code by separating the different views or function.
2) app rout provide flexibility by allowing us to hande different HTTP methods and to add URL parameters to our routes.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() funtion is used for URL building . It accepts the name of function as first argument, and one or more keyword arguments each corresponding to the variable part of URL.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello {name}!'

@app.route('/')
def index():
    return f'The URL for the hello function is {url_for("hello", name="John")}'

if __name__ == '__main__':
    app.run(debug=True)
```
