<h3>Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of 
Matplotlib.

`Matplotlib` is a popular data visualization library in Python. It provides a wide range of functions and tools for creating static, animated, and interactive visualizations in Python. Matplotlib is widely used in various fields such as data analysis, scientific research, and machine learning.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line Plot: This plot is used to visualize the relationship between two continuous variables by connecting data points with straight lines.

2. Bar Plot: It represents data using rectangular bars, where the height or length of the bar corresponds to the data value. It is commonly used to compare different categories or groups.

3. Scatter Plot: This plot is used to display the relationship between two variables as a collection of points. Each point represents an observation, and the position of the point is determined by the values of the two variables.

4. Histogram: A histogram is used to visualize the distribution of a continuous variable. It divides the data into bins and displays the frequency or count of observations falling into each bin.

5. Pie Chart: It represents data as a circular chart divided into slices, where each slice represents a category or group. The size of each slice corresponds to the proportion or percentage of that category in the whole.


<h3>Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.



```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

##### Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
plt.scatter(x=x,y=y,color='g')
plt.xlabel('This is x-axis')
plt.ylabel('This is y-axis')
plt.title("plot for x and y data ")
plt.show
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in a structured manner, making it easier to compare and analyze different visualizations simultaneously.

```python
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Line 1')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Line 2')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Line 3')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Line 4')

plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

`A bar plot,` also known as a bar chart or bar graph, is a visualization that uses rectangular bars to represent categorical or discrete data. It is used to visually compare and display the values of different categories or groups.

`The primary purpose` of a bar plot is to showcase the distribution, comparison, or relationship between different categories or groups. It allows us to quickly interpret and understand the data by visually comparing the heights or lengths of the bars.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
#bar plot
plt.bar(company,profit,color='y')
plt.xlabel('Companies')
plt.ylabel('Profit')

plt.show
```

```python
#horizonal bar graph
plt.barh(company,profit,color='y')
plt.xlabel('Profit')
plt.ylabel('Companies')

plt.show
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

`A Box Plot` is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

`A box plot is used` to summarize and compare distributions, identify outliers, and visualize skewness or symmetry in a dataset.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, ax = plt.subplots()

ax.boxplot([box1, box2])

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')
plt.show()
```
