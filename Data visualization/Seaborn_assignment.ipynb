<h3>Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. `Scatter Plot:` seaborn.scatterplot()
   Use: Shows the relationship between two variables by displaying individual data points as dots on a two-dimensional plane.

2. `Relational Plot:` seaborn.relplot()
    Use: Provides a high-level interface for creating different types of plots, including scatter plots, line plots, and more, based on the specified     parameters.

3. `Joint Plot:` seaborn.jointplot()
    Use: Combines two different univariate or bivariate plots, such as a scatter plot and a histogram, to visualize the relationship between two variables.

4. `Displot:` seaborn.displot()
    Use: Creates a flexible and versatile visualization for the distribution of a single variable or the relationship between two variables.

5. `Categorical Plot:` seaborn.catplot()
    Use: Creates categorical plots that show the relationship between categorical variables and other variables.
